```r
# Ask the user for their name
user_name <- readline(prompt = "Enter your name: ")

# Print a greeting
cat("Hello,", user_name, "! Welcome to R.\n")
```

```r
n_number<-function(n){
  for(i in 2:n){
    cat(i,"\n")
  }
}
n<-as.integer(readline(prompt="Enter"))
n_number(n)
```

```r
n_numbers1<-function(n){
  for(i in n:1){
    cat(i,"\n")
  }
}
n<-as.integer(readline(prompt="Enter number: "))

n_number(n)
```

```r
n<-10
i<-1

while(i<=n){
  cat(i,"\n")
  i<-i+1
}
```

```r
n<-10
i<-n
while(i>=1){
  cat(i,"\n")
  i<-i-1
}
```

```r
letter_vec<-letters

i<-1

while(i<=length(letter_vec)){
  cat(letter_vec[i],"\n")
  i<-i+1
}
```
